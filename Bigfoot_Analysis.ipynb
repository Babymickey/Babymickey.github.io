```python
import altair as alt

# Define the full list of seasons for the color scale domain
seasons = ['Summer', 'Fall', 'Winter', 'Spring', 'Unknown']

# 1. Create the dropdown with 'All' mapped to None
input_dropdown = alt.binding_select(
    options=[None] + seasons, # The actual values in the data (None = show all)
    labels=['All'] + seasons, # What the user sees in the menu
    name='Pick a Season: '
)

selection = alt.selection_point(
    fields=['season'],
    bind=input_dropdown,
    value=None # Start with 'All' selected
)

# 2. Build the chart
bigfoot_chart_inter = alt.Chart(url).mark_circle().encode(
    x=alt.X('longitude:Q').scale(domain=[-130, -60]), # Optional: pin zoom to USA
    y=alt.Y('latitude:Q').scale(domain=[20, 55]),
    color=alt.Color('season:N').scale(domain=seasons), # This keeps colors consistent
    tooltip=['title:N', 'season:N', 'state:N']
).add_params(
    selection
).transform_filter(
    selection
).properties(
    width=600,
    height=400,
    title="Bigfoot Sightings by Season"
)

bigfoot_chart_inter
```

```python
# Save Plot 1 (Static or Simple)
bigfoot_chart.save('bigfoot_chart.json')

# Save Plot 2 (Interactive)
bigfoot_chart_inter.save('bigfoot_chart_inter.json')
```
